## Data Processing

I will build a data processing function to process both datasets for further analysis

```python
def clean_data(df, unknowns=unknowns, recode=recode_CAMEO_DEU_2015, 
                 cols_to_float=cols_to_float, to_binary=to_binary_OST_WEST_KZ, 
                col_for_drop=col_for_drop):
    
    """ 
    Cleaning and imputing dataset  
    
    input: 
    
    df: data frame that we want to be prepared for clustering 
    unknowns: unknown values
    recode: the values for recoding the feature CAMEO_DEU_2015
    cols_to_float: columns to be changed to floats
    to_binary: dictionary for recoding values to binary 
    col_for_drop: columns to be droped
    
    Output:
    df_scalled: the data frame after cleaned, recoded and scalled
    
    """
        
    # Iterating through each coloumn to replace coded unknown values with np.nan
    for col, val in unknowns.items():
        if col in df.columns:
            df[col] = df[col].replace(val, np.nan)
        else:
            continue
            
     # I will recode the attribute 'CAMEO_DEU_2015' to an ordinal catagory
    for key, val in recode.items():
        df['CAMEO_DEU_2015'] = df['CAMEO_DEU_2015'].replace(key, val)
        
    # Change object attributes to float
    df[cols_to_float] = df[cols_to_float].replace(['X', 'XX'], np.nan).astype(float)
    
    # I will change the attribute 'OST_WEST_KZ' to binary
    for key, val in to_binary.items():
        df['OST_WEST_KZ'] = df['OST_WEST_KZ'].replace(key, val)
        
    # Drop the columns with more than 30% values missing
    col_for_remain = list(set(df.columns)-set(col_for_drop))
    df = df.loc[:, col_for_remain]
    
    perc_miss_row = df.isnull().sum(axis=1) / df.shape[1]
    # Data frame of missing values in percentage per row
    miss_val_sorted_row = pd.DataFrame({'perc_miss_row': perc_miss_row}).sort_values('perc_miss_row', ascending=False)
    row_for_drop = miss_val_sorted_row[miss_val_sorted_row['perc_miss_row'] > 0.7].index.tolist()
    df = df.drop(row_for_drop)
    
    # Impute nan values
    imputer = SimpleImputer()
    df_imputed = imputer.fit_transform(df, 'mode')
    df_imputed = pd.DataFrame(df_imputed, columns=df.columns)
    
    return df_imputed
```

```python
def scale_data(df):
    '''
    input:
    df: data for scalling
    
    Output:
    df_scalled: the data frame after cleaned, recoded and scalled
    
    '''

    # Feature scalling 
    scaler = StandardScaler()
    df_scalled = scaler.fit_transform(df)
    df_scalled = pd.DataFrame(df_scalled, columns=df.columns)
    
    return df_scalled
```

## PCA

After cleaning, imputing and scaling the data, I will use princple component analysis to reduce dimention of the original data for unsupervised ML  

```python
# Build a PCA function for reducing dimension.
def pca(n_components, df):
    pca_fit = PCA(n_components).fit(df)
    pca_trans = pca_fit.transform(df)

    return pca_trans, pca_fit
```

```python
# using pca function on the azdias_ready dataset
azdias_trans, azdias_fit = pca(None, azdias_ready)
```

## Clustering

Lets first apply the K-Means clustering to the generel population 

```python
def kmeans(df, ctr, batch_size):
    '''
    Fit the data with Kmeans clustering
    
    INPUT:
    df: data to fit with kmeans clustering
    ctr - k value
    
    OUTPUT:
    score - SSE score
    '''
    k_means = MiniBatchKMeans(n_clusters=ctr, batch_size=batch_size, random_state=42).fit(df)
    model = k_means.fit(df) # fit the model
    score = np.abs(model.score(df)) # get the SSE score
    
    return score
```

```python
# get SSE scores for further visulizasion
scores = []
for ctr in list(range(1,31, 1)):
    scores.append(kmeans(azdias_trans_180, ctr, 40000))
```

```python
plt.plot(list(range(1,31, 1)), scores, linestyle='--', marker='o', color='r')
plt.xlabel('K_value')
plt.ylabel('SSE score')
plt.title('SSE score & K values')
```

```python
def process_data(df, unknowns=unknowns, recode=recode_CAMEO_DEU_2015, 
                 cols_to_float=cols_to_float, to_binary=to_binary_OST_WEST_KZ, 
                col_for_drop=col_for_drop):
    
    """ 
    Cleaning and imputing dataset  
    
    input: 
    
    df: data frame that we want to be prepared for clustering 
    unknowns: unknown values
    recode: the values for recoding the feature CAMEO_DEU_2015
    cols_to_float: columns to be changed to floats
    to_binary: dictionary for recoding values to binary 
    col_for_drop: columns to be droped
    
    Output:
    df_scalled: the data frame after cleaned, recoded and scalled
    
    """
        
    # Iterating through each coloumn to replace coded unknown values with np.nan
    for col, val in unknowns.items():
        if col in df.columns:
            df[col] = df[col].replace(val, np.nan)
        else:
            continue
            
     # I will recode the attribute 'CAMEO_DEU_2015' to an ordinal catagory
    for key, val in recode.items():
        df['CAMEO_DEU_2015'] = df['CAMEO_DEU_2015'].replace(key, val)
        
    # Change object attributes to float
    df[cols_to_float] = df[cols_to_float].replace(['X', 'XX'], np.nan).astype(float)
    
    # I will change the attribute 'OST_WEST_KZ' to binary
    for key, val in to_binary.items():
        df['OST_WEST_KZ'] = df['OST_WEST_KZ'].replace(key, val)
        
    # Drop the columns with more than 30% values missing
    col_for_remain = list(set(df.columns)-set(col_for_drop))
    df = df.loc[:, col_for_remain]
    
    # Impute nan values
    imputer = SimpleImputer()
    df_imputed = imputer.fit_transform(df, 'mode')
    df_imputed = pd.DataFrame(df_imputed, columns=df.columns)
    
    return df_imputed
```

## Supervised Learning Model and Evaluation Metrics 

### Evaluation Metric

Considering that the original dataset is highly inbalanced (1.2%), so ROC_AUC score will be used for model evaluation.

### Model Selection and Evaluation 

Logistic Regression, Random Forest classifier and Gradient Boosting were initially selected as the classifier models for our supervised learning, and they will be evaluated with ROC_AUC score.

### Model training and cross validation

Since the data is unbalanced (1.2%), I decided to use GridSearchCV and cross validate the models with 5 fold split

```python
# Classifiers for evaluate
models = {'LogisticRegression': LogisticRegression(random_state=42),
         'RendomForest': RandomForestClassifier(random_state=42),
         'GradientBoosting': GradientBoostingClassifier(random_state=42)}
```

```python
# Creating pipline
def create_pipeline (classifier):
    pipeline = Pipeline([('scaler', StandardScaler()),
                        ('model', classifier),
    ])
    return pipeline
```

```python
# Camparing three classifier using GridSearchCV
for key, value in models.items(): 
    grid = GridSearchCV(create_pipeline(value), param_grid={}, scoring='roc_auc', cv=5)
    grid.fit(mailout_x, mailout_y)
    print('ROC_AUC score for {} is {}'.format(key, grid.best_score_))
```

### Model tuning using GridSearchCV

GradientBoostingClassifier has the best ROC_AUC score of 0.7, therefore we will select it as our supervised learning models. Next, we will be tuning the model for our final supervised learning model to submit to the KAggle Competition.

```python
# Best tuning the model
model= models['GradientBoosting']
params = {'model__learning_rate': [0.1, 0.2], 'model__n_estimators': [100]}

grid = GridSearchCV(create_pipeline(model), param_grid=params, scoring='roc_auc', cv=5)
grid.fit(mailout_x, mailout_y)
best_model = grid.best_estimator_
print('ROC_AUC score for {} is {}'.format(best_model, grid.best_score_))
print('Best Parameters: {}'.format(grid.best_params_))
```
